### **4.1 Flowchart Dasar: Menentukan Bilangan Ganjil atau Genap**
<img src="1.jpg" width="850px" alt="Deskripsi Gambar" />

---


```python
# Ketikan kode dan penjelasan disini!

angka = int(input("Masukkan sebuah angka: "))
if angka % 2 == 0:
    print(f"{angka} adalah bilangan Genap")
else:
    print(f"{angka} adalah bilangan Ganjil")

# Flowchart
from graphviz import Digraph

dot = Digraph(comment='Flowchart Ganjil Genap')
dot.attr(rankdir='TB', size='8,5')

dot.node('A', 'Mulai')
dot.node('B', 'Input angka')
dot.node('C', 'angka % 2 == 0')
dot.node('D', 'Cetak "Genap"')
dot.node('E', 'Cetak "Ganjil"')
dot.node('F', 'Selesai')

dot.edges(['AB', 'BC'])
dot.edge('C', 'D', 'Ya')
dot.edge('C', 'E', 'Tidak')
dot.edge('D', 'F')
dot.edge('E', 'F')

dot.render('flowchart_ganjil_genap', format='png', view=True)

'''
Pseudocode 

Mulai
Input angka
Jika angka modulus 2 == 0 maka
    Cetak "Bilangan Genap"
Jika tidak
    Cetak "Bilangan Ganjil"
Selesai
'''
```


### **4.2 Visualisasi Data dengan Matplotlib**
<img src="2.jpg" width="650px" alt="Deskripsi Gambar" />

---


```python
# Ketikan kode dan penjelasan disini!

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, label='sin(x)')
plt.xlabel('Waktu')
plt.ylabel('Amplitudo')
plt.title('Visualisasi Fungsi Sinus')
plt.legend()
plt.grid(True)
plt.show()

'''
Penjelasan:
Sumbu x merepresentasikan nilai waktu yang dihasilkan dari np.linspace(0, 10, 100).
Sumbu y merepresentasikan nilai amplitudo fungsi sinus dari setiap nilai x, yaitu hasil dari np.sin(x).
Dengan kata lain, grafik ini menunjukkan bagaimana nilai fungsi sinus berubah terhadap waktu.
'''

# Plot untuk fungsi cos(x)
y_cos = np.cos(x)
plt.plot(x, y_cos, label='cos(x)', color='orange')

# Tampilkan 
plt.legend()
plt.show()
```


### **4.3 Visualisasi Graf Relasi dengan NetworkX**

<img src="3.jpg" width="650px" alt="Deskripsi Gambar" />
---

```python
# Ketikan kode dan penjelasan disini!

import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()
G.add_nodes_from(["A", "B", "C", "D", "E"])
G.add_edges_from([("A", "B"), ("A", "C"), ("B", "D"), ("C", "E"), ("D", "E")])

pos = nx.spring_layout(G)  
nx.draw(G, pos, with_labels=True, node_color='skyblue', node_size=1500)
plt.title("Graf Relasi Entitas")
plt.show()

'''
Penjelasan:
Hubungan antar node pada graf ini terbentuk melalui edge (sisi) yang menghubungkan dua node (simpul).
Misalnya, node "A" terhubung ke node "B" dan "C" karena terdapat edge ("A", "B") dan ("A", "C").
Node "B" terhubung ke "D", "C" terhubung ke "E", dan "D" terhubung ke "E".
Setiap edge merepresentasikan relasi langsung antara dua entitas (node).
Dengan demikian, struktur graf ini menunjukkan bagaimana setiap entitas saling berhubungan secara visual dan matematis.
'''

# Tambahkan dua simpul baru: "F" dan "G"
G.add_nodes_from(["F", "G"])

# Tambahkan tiga sisi baru: ("E", "F"), ("F", "G"), dan ("B", "G")
G.add_edges_from([("E", "F"), ("F", "G"), ("B", "G")])

# Visualisasi
pos = nx.spring_layout(G)  
nx.draw(G, pos, with_labels=True, node_color='lightgreen', node_size=1500)
plt.title("Graf Relasi Entitas (Update)")
plt.show()
```



### **4.4 Flowchart dengan Graphviz: Linear dan Binary Search**
<img src="4.jpg" width="450px" alt="Deskripsi Gambar" />

---

```python
# Ketikan kode dan penjelasan disini!

from graphviz import Digraph

dot = Digraph(comment='Algoritma Pencarian Linear')
dot.node('A', 'Start')
dot.node('B', 'Baca Data Input')
dot.node('C', 'Inisialisasi Indeks = 0')
dot.node('D', 'Indeks < Panjang Data?')
dot.node('E', 'Data[Indeks] == Target?')
dot.node('F', 'Cetak "Ditemukan"')
dot.node('G', 'Indeks += 1')
dot.node('H', 'Cetak "Tidak Ditemukan"')
dot.node('I', 'End')

dot.edges(['AB', 'BC', 'CD', 'DE', 'EG', 'GH', 'HI'])
dot.render('linear_search_flowchart', view=True)

# Modifikasi flowchart untuk algoritma binary search
dot_bin = Digraph(comment='Algoritma Binary Search')
dot_bin.node('A', 'Start')
dot_bin.node('B', 'Baca Data Input (sorted)')
dot_bin.node('C', 'Inisialisasi low=0, high=n-1')
dot_bin.node('D', 'low <= high?')
dot_bin.node('E', 'mid = (low+high)//2')
dot_bin.node('F', 'Data[mid] == Target?')
dot_bin.node('G', 'Cetak "Ditemukan"')
dot_bin.node('H', 'Data[mid] < Target?')
dot_bin.node('I', 'low = mid+1')
dot_bin.node('J', 'high = mid-1')
dot_bin.node('K', 'Cetak "Tidak Ditemukan"')
dot_bin.node('L', 'End')

dot_bin.edges(['AB', 'BC', 'CD'])
dot_bin.edge('D', 'E', 'Ya')
dot_bin.edge('D', 'K', 'Tidak')
dot_bin.edge('E', 'F')
dot_bin.edge('F', 'G', 'Ya')
dot_bin.edge('F', 'H', 'Tidak')
dot_bin.edge('G', 'L')
dot_bin.edge('H', 'I', 'Ya')
dot_bin.edge('H', 'J', 'Tidak')
dot_bin.edge('I', 'D')
dot_bin.edge('J', 'D')
dot_bin.edge('K', 'L')

dot_bin.render('binary_search_flowchart', view=True)
```

<img src="5.jpg" width="650px" alt="Deskripsi Gambar" />

---

```python
# Ketikan kode dan penjelasan disini!

from plantuml import PlantUML
from IPython.display import SVG, display

uml_code = """
@startuml
class Mahasiswa {
    - nim: string
    - nama: string
    + getNIM(): string
    + getNama(): string
}

class MataKuliah {
    - kode: string
    - nama: string
    + getKode(): string
}

Mahasiswa "1" -- "n" MataKuliah : Mengambil
@enduml
"""

server = PlantUML(url='http://www.plantuml.com/plantuml/svg/')
svg = server.processes(uml_code)
display(SVG(svg))

'''
Penjelasan:
Pada diagram UML di atas, terdapat dua kelas: Mahasiswa dan MataKuliah.
Relasi "Mahasiswa '1' -- 'n' MataKuliah : Mengambil" menunjukkan bahwa
satu objek Mahasiswa dapat mengambil banyak MataKuliah (relasi one-to-many).
Artinya, setiap mahasiswa bisa terdaftar di beberapa mata kuliah,
sedangkan satu mata kuliah bisa diambil oleh banyak mahasiswa (relasi many-to-many jika dilihat dari kedua arah).
'''
```
